```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
dx=pd.read_csv('Linear_X_Train.csv')
dy=pd.read_csv('Linear_Y_Train.csv')

x=dx.values
y=dy.values

x=x.reshape((-1,1))
y=y.reshape((-1,1))

X = (x-x.mean())/x.std()
Y = y



```

```python
plt.scatter(X,Y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

```

```python
model = LinearRegression()
```

```python
model.fit(X,Y)
```

```python
output = model.predict(X)
```

```python
bias = model.intercept_
coeff = model.coef_

print(bias)
print(coeff)
```

```python
model.score(X,Y)
```

```python
plt.scatter(X,Y,label='data')
plt.plot(X,output,color='black',label='prediction')
plt.legend()
plt.show()
```
