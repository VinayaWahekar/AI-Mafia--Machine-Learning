```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
col=['feature_1','feature_2','feature_3','feature_4','feature_5','target']
dx=pd.read_csv('Train.csv',usecols=col)


X=dx[['feature_1','feature_2','feature_3','feature_4','feature_5']]
Y=dx['target']


print(X.shape)
print(Y.shape)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X.feature_1,Y,label='feature1',color='green')
plt.scatter(X.feature_2,Y,label='feature2',color='red')
plt.scatter(X.feature_3,Y,label='feature3',color='blue')
plt.scatter(X.feature_4,Y,label='feature4',color='violet')
plt.scatter(X.feature_5,Y,label='feature5',color='yellow')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,Y)
```

```python
output = model.predict(X)
```

```python
bias = model.intercept_
coeff = model.coef_

print(bias)
print(coeff)
```

```python
model.score(X,Y)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(X.feature_1,Y,label='data')
plt.scatter(X.feature_2,Y,label='data')
plt.scatter(X.feature_3,Y,label='data')
plt.scatter(X.feature_4,Y,label='data')
plt.scatter(X.feature_5,Y,label='data')
plt.plot(X.feature_1,output,color='orange',label='feature 1')
plt.plot(X.feature_2,output,color='yellow',label='feature 2')
plt.plot(X.feature_3,output,color='red',label='feature 3')
plt.plot(X.feature_4,output,color='magenta',label='feature 4')
plt.plot(X.feature_5,output,color='cyan',label='feature 5')
plt.legend()
plt.show()
```

```python

col=['feature_1','feature_2','feature_3','feature_4','feature_5']
dxtest=pd.read_csv('Test.csv',usecols=col)


dX=dxtest[['feature_1','feature_2','feature_3','feature_4','feature_5']]

```

```python
output_test=model.predict(dX)
```

```python

plt.scatter(dX.feature_1,output_test,label='feature1',color='green')
plt.scatter(dX.feature_2,output_test,label='feature2',color='red')
plt.scatter(dX.feature_3,output_test,label='feature3',color='blue')
plt.scatter(dX.feature_4,output_test,label='feature4',color='violet')
plt.scatter(dX.feature_5,output_test,label='feature5',color='yellow')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.plot(dX.feature_1,output_test,color='orange',label='feature 1')
plt.plot(dX.feature_2,output_test,color='yellow',label='feature 2')
plt.plot(dX.feature_3,output_test,color='red',label='feature 3')
plt.plot(dX.feature_4,output_test,color='magenta',label='feature 4')
plt.plot(dX.feature_5,output_test,color='cyan',label='feature 5')
plt.legend()
plt.show()
```
